# XLSX

XLSX: es el formato estándar de Microsoft Excel, que es el conocido programa de hojas de cálculo de Microsof


##Leer datos
La función `read_excel()` lee un archivo de Excel en un pandas DataFrame (aunque también puede leer archivos xls y odt). Dado que los archivos de Excel pueden contener varias hojas, esta función puede leerlas todas.

Dentro de los parámetros aceptados por esta función se encuentran:
*   `sheet_name`: Puede ser `str`, `int`, `list` o `None`. El valor predeterminado es 0. Dado que los archivos de Excel se componen de hojas, se puede especificar una sola hoja o varias para leer. Las hojas se pueden especificar con números enteros `[1, 3, 7]` o con el nombre `["Sheet4", "Sheet5", "Sheet8"]`.
*   `header`: la fila que debe considerarse como encabezado del archivo de Excel. El valor predeterminado es `0` porque la primera fila suele ser el encabezado y considera la indexación basada en cero. Si no hay encabezado, el valor debe de ser `None`.
*   `names`: Columnas que se desean leer del archivo de Excel. Acepta una lista de nombres de columna. El valor predeterminado es `None`.

# Archivos csv

CSV (*Comma Separated Values*): como su nombre lo indica, contienen valores separados por comas (,), aunque también se puede usar el punto y coma (;) como separador. Esto genera una especie de tabla en filas y columnas, la cual se puede leer en cualquier editor de texto, e incluso en excel



##Leer datos

Para leer datos, se utiliza el comando `pd.read_csv`, el cual toma por defecto el separador de ",".

```python
Frame=pd.read_csv('EjemploDatosCSV.csv')
Frame.head()
```

En caso de que el documento utilice ";" para separar los datos, es necesario especificarlo con: `pd.read_csv(Archivo,sep=';')`

```python
Frame=pd.read_csv('EjemploDatosCSV2.csv', delimiter=';')
```

También existen casos en los que los datos no tienen un encabezado (nombre de la columna), por lo que se pueden agregar una vez leídos los archivos, esto se hace con el parámetro `names`. En este tipo de lectura, es importante agregar `header=None`, por lo que no leerá la primera fila de datos como encabezado.

```python
Frame=pd.read_csv('EjemploDatosCSV3.csv')
```
