<h1>Agregar columnas (20 puntos)</h1>

<p>5. Se deben calcular otros datos a fin de tener más información de los productos. Por lo tanto,
se deben agregar las siguientes columnas al DataFrame:<br>
• ‘PrecioVenta’: El precio al que se venden los productos es el 30% más del precio de
compra.<br>
• ‘GananciaInventario’: Ganancia en caso de que se venda toda la existencia del
producto (OJO: Ganancia, no total de la venta)
Imprime los últimos 5 registros del frame para comprobar.<br>
</p>

```python
df['PrecioVenta']=df['PrecioCompra']*1.3
df['GananciaInventario']=(df['PrecioVenta']-df['PrecioCompra'])*df['Existencia']
df.tail(5)
```

<p>6. También se necesita una columna con un valor de ‘Si’ o ‘No’ (con el nombre que prefieras),
dependiendo si se necesita surtir el producto. La papelería considera que un producto se
debe de surtir cuando se tienen 40 o menos unidades de ese producto en
existencia. Imprime los primeros 5 registros del frame para comprobar.</p>

```python
df['Surtir?']=df['Existencia']<=40
df['Surtir?']=df['Surtir?'].map({True:'Si',False:'No'})
df
```

<h1>Operaciones básicas (10 puntos)</h1>

<p>7. Se necesitan obtener los siguientes datos del DataFrame. Obtén lo solicitado, cuidando que
se visualice el resultado obtenido<br><br>
• El total de unidades en existencia (No cuántos productos distintos hay, sino cuántas
unidades se tienen en total, toma en cuenta la existencia por producto).<br>
• Total de productos distintos.<br>
• Promedio de precio de venta.<br>
• Total de ganancias en caso de que se venda todo el inventario</p>

```python
total_existencias=df['Existencia'].sum()
print('Total de existencias:',total_existencias)
total_productos=df['Nombre'].value_counts().sum()
print('Total de productos distintos:',total_productos)
promedio_precio_venta=df['PrecioVenta'].mean().round(2)
print('Promedio de precio de venta:',promedio_precio_venta)
total_ganancias=df['GananciaInventario'].sum()
print('Total de ganancias en caso de que se venda todo el inventario:',total_ganancias)
```
