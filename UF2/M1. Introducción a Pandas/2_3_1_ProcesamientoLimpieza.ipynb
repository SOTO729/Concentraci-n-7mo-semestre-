#Datos faltantes
Se les llama "Datos faltantes" a aquellas "celdas" que no tienen datos o tienen un dato nulo (celda vacía o valor nan).

##Filtrado
Para detectar los datos faltantes, se tienen dos funciones que retornan uns serie de valores lógicos:

`isnull()`: Retorna un True si el dato es nulo y False en caso de que tenga un valor asignado

`notnull()`: Retorna False si es un dato nulo y True si tiene un valor asignado

```python
print(Ej.isnull())
```

```python
print(len(Ej)-Ej.count())
```

##Eliminar datos nulos
`dropna()` retorna una columna o un frame sin los datos nulos. En caso de que se aplique a todo el frame, eliminará una fila si hay por lo menos una columna por un nulo.

**NO ES LO MISMO UN DATO NULO QUE UN 0**

```python
Ej.dropna()
```

```python
Ej['GNPOld'].dropna()
```

```python
print("Sin datos nulos en la columna GNPOld y Name:\n",Ej[['GNPOld','Name']].dropna())
```

```python
print("Numero de elementos nulos (IndepYear):",Ej['IndepYear'].dropna().count())
```

dropna tiene dos parámetros que modifican su funcionamiento:
 -- `axis`: para especificar si los valores nulos se buscarán por fila (`0`) o columna (`1`)

 -- `how`: Si borra el elemento con un solo valor nulo (`'any'`) o todos los elementos deben de ser nulos (`'all'`)

```python
#Borra cualquier fila que tenga un valor nulo
print("\n\n\nAxis=0, How=Any:\n",Ej.dropna(axis=0, how='any'))
```

```python
#Se cambian los valores a NaN de una fila y columna para ver el funcionamiento de how='all'
from numpy import nan
Ej['Name']=nan
Ej.loc['AND']=nan
Ej
```

```python
#Borra cualquier fila donde TODOS sus valores sean nulos
print("\n\n\nAxis=0, How=all:\n",Ej.dropna(axis=0, how='all'))
```

```python
#Borra cualquier columna donde TODOS sus valores sean nulos
print("\n\n\nAxis=1, How=all:\n",Ej.dropna(axis=1, how='all'))
```

##Modificar el valor de los datos nulos
Se utiliza la función `fillna` para reemplazar los valores nulos por un dato específico

```python
Ej=pd.read_csv('Country2.csv', index_col='Code', nrows=20)
Ej=Ej.drop(['SurfaceArea','LifeExpectancy','Continent','Region','LocalName','GovernmentForm','Capital'], axis=1)
Ej.head()
```

```python
Ej.fillna('--',inplace=True)
Ej
```

```python
Ej.fillna({'Name':'ND','IndepYear':int(Ej['IndepYear'].mean()),'GNPOld':int(Ej['GNPOld'].mean())})
```

```python
#Para reemplazar los datos (0) por otro valor
print("Reemplaza todos los 0's por 9999\n",Ej.replace(0,9999))
```
