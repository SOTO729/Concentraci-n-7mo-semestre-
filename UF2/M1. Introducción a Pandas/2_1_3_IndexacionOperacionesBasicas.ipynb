#Indexar / reindexar
**Indexar:** Cambiar el nombre de las claves, ya sea de filas o columnas

**Reindexar:** Cambiar de posición las claves (filas o columnas)

##Series

```python
import pandas as pd
from pandas import Series
from pandas import DataFrame
s=pd.Series([45,32,21,34,45,65],['a','s','d','e','t','y'])
s
```

```python
#Modificar los índices
print(s.index)
s.index=['q','i','o','u','r','y']
s
```

```python
#Reajustar los índices
#No cambia valores, solo posiciones
s=s.reindex(['o','i','y','r','u','q'])
s
```

```python
#Cambiar un índice específico
s=s.rename({'y':'x'})
s=s.rename({'o':'w','i':'b'})
s
```

##Ordenar

###Series
Para ordenar series, utilizaremos las funciones:

`sort_index` para ordenar por la clave

`sort_values` para ordenar de acuerdo al valor

Como parámetros utilizaremos `ascending` para determinar el orden (ascendente o descendente) e `inplace` para determinar si se modifica la serie o solo se imprime

```python
import pandas as pd
from pandas import Series
s=pd.Series([45,32,21,34,45,65],['a','s','d','e','t','y'])
s
```

```python
#Por defecto no modifica la serie y ordena ascendentemente
s.sort_index()
```

```python
s.sort_index(ascending=False)
```

```python
s.sort_index(ascending=False).tail(3)
```

```python
s.sort_values()[:3]
```

```python
s.sort_values().head(3)
```

```python
#Mostrar los primeros 4 valores despues de ordenar los indices en orden descendente
s.sort_index(ascending=False).head(4)
```

###Frames
Para los frames, se utilizan los mismos parámetros que en las series, pero se agrega:

`axis:` determinar si es por filas (1) o columnas (0) en el caso de ordenar índices

`by:` En caso de que sea por columnas, cuál columna se tomará como referencia

```python
from pandas import DataFrame
data = {'Sucursal': ['Independencia', 'Reforma', 'Centro', 'Periférico', 'Forjadores', 'Bicentenario','Universidad'],
'Gerente':['Javier Méndez','Susana Reyes','Alberto Suez','Elmer Figueroa','Elena Pérez','Juan Avendaño','René Suarez'],
'Enero': [1800,12344,23000,45000,3400,23000,5600],
'Febrero': [2000,10000,26000,50000,5000,20000,5000],
'Marzo':[1000,15000,23000,48000,6000,18000,6000]}

df=pd.DataFrame(data, index=['ab','ac','ad','ca','bc','bd','aa'])
df['Categoria predominante']=['Electrónica','Herramienta','Herramienta','Herramienta','Electrónica','Electrónica','Herramienta']
df
```

```python
#Ordenar de acuerdo a las claves
df.sort_index()
df
```

```python
#De acuerdo a los valores
df.sort_values(by='Categoria predominante',ascending=False)
df
```

```python
#                   Ordena de izquierda a derecha!!!!
df.sort_values(by=['Categoria predominante','Enero'],ascending=False) #
df
```

```python
#Ordenar por varias columnas, pero en diferente orden
#Categoria --> ascendente
#Enero ---> descendente
df.sort_values(by=['Categoria predominante','Enero'],ascending=[True,False])
df
```

```python
df.sort_values(by='Enero',ascending=False)
df
```

##Columnas derivadas

Permiten generar nuevas columnas en relación a una condicional. Para esto, es necario importar la librería numpy

```python
df
```

```python
import numpy as np
#Where
#Si las ventas de enero son mayores a 20000, poner "Buenas", si no, poner "Malas"
df['Ventas']=np.where(df['Enero']>20000,"Buenas","Malas")
df
```

```python
df['Ventas']=np.where(df['Enero']>20000,"Buenas",np.where(df['Enero']>15000,'verde',np.where(df['Enero']>5000,'azul','rojo')))
df
```

```python
#Columnas derivadas
import numpy as np


#Columna "tipo" con valor de "básica" si las ventas totales son mayores de 50,000 y "quiebra" en caso contrario
df['Tipo'] = np.where(df['Enero']+df['Febrero']+ df['Marzo']>50000, "Básico", "Quiebra")
```

#Filtros

Los fitros ayudan a mostra información que cumple con ciertas características. Para esto se utilizan condicionales en el siguiente formato

`frame[columnas_a_mostrar][Condicionales]`

```python
#Ver todas las columnas de las sucursales con ventas regulares
print(df[df['Ventas']=="Regulares"])
```
