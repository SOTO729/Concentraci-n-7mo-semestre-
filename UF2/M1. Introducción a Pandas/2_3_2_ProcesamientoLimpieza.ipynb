#Duplicados
Se utiliza la función `duplicated()`,la cual retorna un valor booleano en caso de que el dato esté duplicado en filas anteriores.

```python
import pandas as pd
from pandas import DataFrame

Ej=pd.read_csv('EjemploDatosCSV.csv', index_col='Clave')
Ej=Ej.drop(['Apellido','Cargo','Fecha ingreso'], axis=1)
Ej
```

```python
#Se agregan dos elementos duplicados (Con distinta clave) para probar la función
Ej.loc['FerRt-45']=Ej.loc['GarFe-36']
Ej.loc['AntMa-34']=Ej.loc['ArtAn-51']
Ej.loc['AbcDf-78']=Ej.loc['ArtAn-51']
Ej
```

```python
print("Filas duplicadas:\n",Ej.duplicated())
```

```python
Ej['Departamento'].duplicated(keep='first')
```

```python
Duplicadas=Ej.duplicated()
Duplicadas
```

```python
print("\n\nNúmero de filas duplicadas:\n", Duplicadas[Duplicadas==True].count())
```

```python
print(Ej['Departamento'])
```

La función duplicated tiene el parámetro `keep` para determinar si mantiene el primer registro (`'last'`), el último (`'last'`) o ninguno (`False`) de los duplicados

```python
#Valores duplicados de una columna:
Ej['Departamento'].duplicated(keep='last')
```

```python
"""¿Cuántos departamentos existen?"""
NoDepto=Ej['Departamento'].duplicated(keep='last')
print('Total Depto',NoDepto[NoDepto==False].count())
```

```python
No=Ej['Sueldo'].duplicated(keep=False)
print('Numero de sueldos únicos',NoDepto[No==False].count())
```

```python
Ej['Nombre'][Ej['Sueldo'].duplicated(keep=False)==False]
```

```python
Ej['Sueldo'].duplicated(keep=False)
```

#Outliers
Son datos atípicos, por lo que no se comportan como el resto de ellos. Puede haber tres tipo de outliers

1.- Dato mal capurado

2.- Dato en el límite (es real pero no me interesa analizarlo)

3.- Anómalo Muy curioso


```python
import pandas as pd
from pandas import DataFrame

Ej=pd.read_csv('EjemploDatosCSV.csv', index_col='Clave')
Ej=Ej.drop(['Apellido','Cargo','Fecha ingreso'], axis=1)
Ej
```

```python
Ej.loc['MarLo-59','Sueldo']=-500000
Ej.loc['PerNa-53','Sueldo']=1000000

Ej['Sueldo_Z']= (Ej['Sueldo'] - Ej['Sueldo'].mean()) / Ej['Sueldo'] .std()
print(Ej['Sueldo'].mean(),Ej['Sueldo'] .std())
Ej.sort_values(by='Sueldo')
```

```python
Borrar=Ej[~Ej['Sueldo_Z'].between(-3,3)]
print(Borrar)
Ej.drop(Borrar.index, inplace=True)
Ej
```

```python
Ej['Sueldo_Z']= (Ej['Sueldo'] - Ej['Sueldo'].mean()) / Ej['Sueldo'] .std()
print(Ej['Sueldo'].mean(),Ej['Sueldo'] .std())
Ej.sort_values(by='Sueldo')
```
