# Archivos HTML

HTML (*HyperText Markup Language*) es el código que se utiliza para estructurar y desplegar una página web y sus contenidos.

Una tabla (`<table>`) es un conjunto de celdas normales `<td>` o de tipo encabezado `<th>` organizadas en filas (`<tr>`). Esto permite visualizar una tabla en un archivo .html

image.png

```python
import pandas as pd
from pandas import DataFrame
```

```python
Frame=pd.read_html('Ejemplo-1.html',header=0, index_col=0,encoding='utf-8')
#al leer un archivo html se retorna una lista de frames
Frame[0]#frame es una lista de un elemento (DataFrame)
Frame[0].head()
for x in Frame:
  print(len(x))
  print(x.info())
```

```python
Lista = [3,2,6,8,9,1,0]
for x in Lista: #for es un ciclo
  print(x)
```

```python
Frame=pd.read_html('Ejemplo-3.html',header=0, index_col=0,encoding='utf-8')
len(Frame)
print(Frame[0].head())
print(Frame[1].head())

#imprime el número de filas de cada frame en el archivo (no sabemos cuantos frames existen)
len(Frame)
```

# Archivos JSON

**JSON (Short for JavaScript Object Notation)**: Formato para el intercambio de datos basado en texto, por lo que es fácil de leer (personas y máquina). Los datos en los archivos JSON son pares de propiedad-valor separados por dos puntos.

450_1000.png

Elementos básicos:

*   Llaves { } sirven para delimitar los objetos y son obligatorias para iniciar y terminar el contenido.
*   Corchetes [ ] se usan para indicar un array.
*   Dos puntos : sirven para separar la clave y su valor correspondiente.
*   Coma , se usa para indicar la separación entre los elementos.

##Lectura y edición

Al igual que los archivos CSV y XLSX, se utilizan las funciones `read_json` y `to_json` para leer y pasar al formato JSON respectivamente. En estas funciones, es importante establecer el formato específico que se quiere lograr o que tiene el archivo a leer.


image.png

```python
#Leer un documento json
#values no cuenta con nombre de columnas ni claves asignadas (numeración)
Frame= pd.read_json('Arch5_values.json', orient='values')
Frame.columns=['Nombre', 'Apellido', 'Depto', 'Puesto', 'Fecha', 'Sueldo']
Frame.head()
```

```python
#al leer se debe poner el formato en el que estan los datos
Frame= pd.read_json('Arch3_index.json', orient='columns')
Frame.head()
```

```python
#Se puede escribir en un documento json con el mismo formato, diferente, incluso en un formato CSV o Excel
Frame.to_json('Ejemplo_JSON.json', orient='values', force_ascii=False)
```
