### Segmentando una columna con etiquetas

```python
energy_bins = [8, 35, 45, 64, 80, 97]

#Bin labels must be one fewer than the number of bin edges
energy_labels = ['pobre energía', 'baja energía','moderada energía', 'buena energía', 'alta energía']
```

```python
df['energy_bin'] = pd.cut(df['energy_%'], energy_bins, labels=energy_labels)
```

```python
df[['energy_%', 'energy_bin']].head(10)
```

```python
df[df['energy_%']< 20][['energy_%', 'energy_bin']].head(10)
```

```python
df[df['energy_%']< 20]
```

### Segmentando una etiqueta por cuantiles

```python
n_quantils = 5
```

```python
df['streams_q_bin'] = pd.qcut(df['streams'], q=n_quantils, labels=False)
```

```python
df[['streams','streams_q_bin']]
```

```python
df['streams'].mean()
```

```python
df[df['streams_q_bin'] == 2][['streams','streams_q_bin']]
```

```python
for quantil in range(0,n_quantils):
    min_value = df[df['streams_q_bin'] == quantil]['streams'].min()
    max_value = df[df['streams_q_bin'] == quantil]['streams'].max()
    print(f"{quantil}: [{min_value}, {max_value}]")
```

```python
df['streams_q_bin'].hist()
```

```python
df['streams'].hist()
```

```python
labels= ['bajo', 'medio', 'alto', 'muy alto', 'extremadamente alto']
df['streams_q_bin'] = pd.qcut(df['streams'], q=n_quantils, labels=labels)
```

```python
df[['streams','streams_q_bin']]
```

```python
df[df['streams_q_bin']  == 'alto']['streams'].count()
```

```python
df[df['streams_q_bin']  == 'muy alto']['streams'].count()
```

### Ordinal encoder (Codificación ordinaria)

```python
df["key"].unique()
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Creamos el codificador indicandole el orden de la variables
encoder = OrdinalEncoder(categories=[["C","C#","D","D#","E","F","F#","G","G#","A", "A#", "B", "C"]])
```

```python
df = df.dropna()
```

```python
encoder.fit(df[["key"]])
df["key-encoded"] = encoder.transform(df[["key"]])
```

```python
df[["key", "key-encoded"]]
```
