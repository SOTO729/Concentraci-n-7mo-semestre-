```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
brkb=pd.read_csv('BRK-B.csv')
brkb
```

```python
xom=pd.read_csv('XOM.csv')
xom
```

```python
stocks=pd.merge(left=brkb,right=xom,left_on='Date',right_on='Date')
stocks
```

```python
stocks[['BRK-B','XOM']]=scaler.fit_transform(stocks[['Close_x','Close_y']])
stocks
```

```python
stocks[['BRK-B','XOM']].plot(title='BRK-B vs XOM',xlabel='Days',ylabel='Price (USD)')
```
