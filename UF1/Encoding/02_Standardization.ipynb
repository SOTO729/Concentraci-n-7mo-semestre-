### Normalización o escala min-max

```python
df[['in_spotify_charts', 'streams']] = (df[['in_spotify_charts', 'streams']] - df[['in_spotify_charts', 'streams']].min()) / (df[['in_spotify_charts', 'streams']].max() - df[['in_spotify_charts', 'streams']].min())
```

```python
df.describe()
```

```python
q1 = df[['in_apple_playlists']].quantile(0.25)
q3 = df[['in_apple_playlists']].quantile(0.75)
iqr = q3 - q1
df[['in_apple_playlists']] = (df[['in_apple_playlists']] - df[['in_apple_playlists']].median()) / iqr
```
